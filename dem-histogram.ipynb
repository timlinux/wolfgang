```python
import rasterio
import matplotlib.pyplot as plt

# Function to open and convert GeoTIFF file to a NumPy matrix
def geotiff_to_matrix(file_path):
    with rasterio.open(file_path) as src:
        matrix = src.read(1)  # Assuming a single-band raster
    return matrix

# Prompt user for the path to the GeoTIFF file
geotiff_path = input("Enter the path to the GeoTIFF file: ")

# Convert GeoTIFF to NumPy matrix
dem_matrix = geotiff_to_matrix(geotiff_path)

# Print basic information about the DEM
with rasterio.open(geotiff_path) as src:
    print(f"Raster Shape: {src.shape}")
    print(f"Number of Bands: {src.count}")
    print(f"Coordinate Reference System (CRS): {src.crs}")

# Plot histogram of DEM values
plt.figure(figsize=(10, 6))
plt.hist(dem_matrix.flatten(), bins=100, color='blue', alpha=0.7)
plt.title('Distribution of DEM Values')
plt.xlabel('Elevation')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```
